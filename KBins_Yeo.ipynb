```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('/content/train.csv',usecols=[1,5,9,])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
dt = DecisionTreeClassifier()

dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test)*100)
```

```python
print(np.mean(cross_val_score(dt,x_train,y_train,cv=5,scoring='accuracy')*100))
```

```python
trf = ColumnTransformer([
    ('kb_age',KBinsDiscretizer(n_bins=15,strategy='quantile',encode='ordinal'),[0]),
    ('kb_fare',KBinsDiscretizer(n_bins=5,strategy='quantile',encode='ordinal'),[1])])
```

```python
trf.named_transformers_['kb_age'].n_bins
```

```python
x_train1 = trf.fit_transform(x_train)

x_test1 = trf.transform(x_test)
```

```python
dt1 = DecisionTreeClassifier()

dt1.fit(x_train1,y_train)

y_pred1 = dt1.predict(x_test1)
```

```python
print(accuracy_score(y_pred1,y_test)*100)
```
