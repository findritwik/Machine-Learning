```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/train.csv')[['Age','Pclass','Fare','Survived']]
```

```python
df
```

```python
df.info()
```

```python
df.isnull().mean()*100
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
si = SimpleImputer()
x_train_trf = si.fit_transform(x_train)
x_test_trf = si.transform(x_test)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train_trf,y_train)
```

```python
y_pred = lr.predict(x_test_trf)
```

```python
accuracy_score(y_pred,y_test)
```

```python
si1 = KNNImputer(n_neighbors=3,weights='distance')
x_train_trf1 = si1.fit_transform(x_train)
x_test_trf1= si1.transform(x_test)
```

```python
lr.fit(x_train_trf1,y_train)
```

```python
y_pred1 = lr.predict(x_test_trf1)
```

```python
accuracy_score(y_pred1,y_test)
```
