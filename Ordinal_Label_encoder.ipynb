```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('/content/customer.csv',usecols=[2,3,4])
```

```python
df
```

```python
df.head()
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

oe.fit(df[['review','education']])

df[['review','education']] = oe.transform(df[['review','education']])
```

```python
oe.categories_
```

```python
le = LabelEncoder()

df['purchased'] = le.fit_transform(df['purchased'])
```

```python
le.classes_
```

```python
scaler = StandardScaler()

scaler.fit(df[['review','education']])

df[['review','education']] = scaler.transform(df[['review','education']])

```

```python
x = df.drop('purchased',axis=1)
y = df['purchased']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y
```

```python
lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
y_predict =lr.predict(x_test)
```

```python
accuracy_score = accuracy_score(y_test,y_predict)
```

```python
print("Score:",accuracy_score)
```

```python
print(cross_val_score(lr,x_train,y_train,cv=10,scoring='accuracy').mean()*100)
```
