```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import Binarizer
```

```python
df = pd.read_csv('/content/train.csv')[['Age','Fare','SibSp','Parch','Survived']]
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df
```

```python
df['Family'] = df['Parch'] + df['SibSp']
```

```python
df
```

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
```

```python
df
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
dt = DecisionTreeClassifier()

dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test)*100)
```

```python
print(np.mean(cross_val_score(dt,x_train,y_train,cv=10,scoring='accuracy')*100))
```

```python
trf = Binarizer(copy=True)


```

```python
trf
```

```python
x_train1 = trf.fit_transform(x_train)

x_test1 = trf.transform(x_test)
```

```python
x_train1
```

```python
x_train1 = pd.DataFrame(x_train1,columns=x_train.columns)
x_test1 = pd.DataFrame(x_test1,columns=x_test.columns)
```

```python
x_train1
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train1,y_train)
y_pred2 = clf.predict(x_test1)

accuracy_score(y_test,y_pred2)
```

```python
np.mean(cross_val_score(clf,x_train1,y_train,cv=10,scoring='accuracy')*100)
```
