```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.compose import ColumnTransformer
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = pd.read_csv('/content/train (1).csv',usecols=['Age','Fare','Survived'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
dt = DecisionTreeClassifier()

dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
cross_val_score(dt,x,y,cv=5,scoring='accuracy').mean()
```

```python
ct = ColumnTransformer([('kb',KBinsDiscretizer(n_bins=15,encode='ordinal',strategy='quantile'),[0,1])],remainder='passthrough')
```

```python
ct.fit(x_train)

x_train1 = ct.transform(x_train)
x_test1 = ct.transform(x_test)
```

```python
dt1 = DecisionTreeClassifier()

dt1.fit(x_train1,y_train)

y_pred = dt1.predict(x_test1)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
np.mean(cross_val_score(dt1,x,y,cv=5,scoring='accuracy'))
```
